# Collecting reviews from Amazon Product Page

### Below functions take url for amazon reviews 1st page as input, collect 1000 reviews and saves them as csv file

```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function 1: to get and return parsed HTML
def get_soup(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    return soup

# Initialize list to store reviews data later on
reviewlist = []

# Function 2: that sifts through the soup element above, tries to find the tags for the data we want, then appends to the list we initialized
def get_reviews(soup):
    reviews = soup.find_all('div', {'data-hook': 'review'})
    try:
        for item in reviews:
            review = item.find('span', {'data-hook': 'review-body'}).text.strip()
            if review == '':
                print('This review is empty')
                pass
            else:
                reviewlist.append(review)
            
    except:
        pass

# Function 3: loop through 1:x many pages, or until the css selector found only on the last page is found (when the next page button is greyed)
for x in range(1,200):
    # replace the url in below statement
    soup = get_soup(f'https://www.amazon.in/OnePlus-Nord-Blue-128GB-Storage/product-reviews/B097RD2JX8/ref=cm_cr_getr_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={x}')
    print(f'Getting page: {x}')
    get_reviews(soup)
    print(len(reviewlist))
    if len(reviewlist) <= 1000:
        if not soup.find('li', {'class': 'a-disabled a-last'}):
            pass
        else:
            break
    else:
        break

# Save results to a dataframe, then export as CSV
df = pd.DataFrame(reviewlist)
df.to_csv(r'reviews.csv', index=False)
print('completed')
```

Above code can be refactored as per need to change the url, number of pages to scrape and number of reviews to save.
